# Cupel — quickstart in Colab

**Cupel** is independent behavioral assurance for an AML transaction-monitoring agent: it quantifies under-escalation that standard observability misses, and emits an interpretable, ground-truth-validated verification record for every decision.

This notebook runs the reference flow end to end. The data layer is **free and offline**; the steps that call a model need your own `ANTHROPIC_API_KEY` and cost a few dollars (each cell states its cost).

Repo: https://github.com/burnssa/cupel  ·  BYO guide: [docs/BYO_GUIDE.md](https://github.com/burnssa/cupel/blob/master/docs/BYO_GUIDE.md)
